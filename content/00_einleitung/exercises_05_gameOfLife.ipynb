## Conway's Game of Life

# Aufgabe 1 - Die Grundregeln

Conway's Game of Life basiert auf wenigen simplen Regeln. Es gibt ein Spielfeld mit einem quadratischen Rastern von Feldern. In jedem Feld kann sich eine Zelle befinden, die unter bestimmten Umständen ans Leben kommt, oder stirbt.

 - Hat eine lebende Zelle genau zwei, oder drei lebende Nachbarn fühlt sie sich wohl und bleibt am Leben.
 
 - Hat eine tote Zelle genau drei lebende Nachbarn wird sie geboren.
 
Implementieren Sie eine Funktion, welche diese Regeln für die angegebene Anzahl an Iterationen auf eine eingegebene Startverteilung von Zellen anwendet und anzeigt. Beachten Sie dabei, dass die Zellen alle gleichzeitig leben, und sich somit auch gleichzeitig ihre Zustände verändern. Der Rand des Spielfeldes gilt dabei als tote Zellen.
