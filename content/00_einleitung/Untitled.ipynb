## Conway's Game of Life

# Aufgabe 1 - Die Grundregeln

Conway's Game of Life basiert auf wenigen simplen Regeln. Es gibt ein Spielfeld mit einem quadratischen Rastern von Feldern. In jedem Feld kann sich eine Zelle befinden, die unter bestimmten Umständen ans Leben kommt, oder stirbt.

 - Hat eine lebende Zelle mehr als drei lebende Nachbarn stirbt sie an Überbevölkerung
 
 - Hat eine Zelle weniger als zwei lebende Nachbarn stirbt sie vor Einsamkeit
 
 - Hat eine tote Zelle genau drei lebende Nachbarn wird sie geboren
 
 Implementieren Sie das Spielfeld und wenden Sie diese Regeln auf alle Zellen an. Denken Sie daran, dass alle Veränderungen gleichzeitig stattfinden müssen. Versuchen Sie außerdem ein Abbruchkriterium zu finden, mit dem die Simulation zu einem sinnvollen Zeitpunkt beendet wird. Achten Sie außerdem auf die implementierung des Verhaltens der Zellen in den Randbereichen, wiederholt sich ihre Welt, oder gibt es einen einsamen, lebensfeindlichen Rand?
